```python
import datetime
import yfinance as yf
import time
```

```python
def get_stock_price():
    start = datetime.datetime(2017, 1, 1).date()
    end = datetime.datetime(2017, 12, 31).date()
    # end = start + datetime.timedelta(days=35)
    # print(end)
    stock_data = yf.download("^HSI", start, end)
    # print(stock_data)
    return stock_data

HSI_price = get_stock_price()
HSI_price
```

```python
cur_day = '01-01-2017'
end_day = '31-12-2017'
cur_day = datetime.datetime.strptime(cur_day, '%d-%m-%Y').date()
end_day = datetime.datetime.strptime(end_day, '%d-%m-%Y').date()
cur_day
```

```python
list_trading_day = [day.date() for day in HSI_price.index.tolist()]
```

```python
holidays = []
while cur_day <= end_day:
    if cur_day not in list_trading_day:
        print(cur_day.strftime('%d-%m-%Y'))
        holidays.append(cur_day.strftime('%d-%m-%Y'))
    cur_day += datetime.timedelta(days=1)
```
