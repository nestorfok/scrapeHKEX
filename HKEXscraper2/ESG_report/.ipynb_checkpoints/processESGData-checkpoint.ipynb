```python
def concatenateData(df_result, df_year, columns):
    for index, row in df_year.iterrows():
        if row['stock_code'] not in df_result['stock_code'].values:
            each_row = {
                'stock_code': [row['stock_code']], 
                'ESG_2017_rel_date': [row['ESG_2017_rel_date']],
                'ESG_2018_rel_date': [row['ESG_2018_rel_date']],
                'ESG_2019_rel_date': [row['ESG_2019_rel_date']],
                'ESG_2020_rel_date': [row['ESG_2020_rel_date']],
                'ESG_2021_rel_date': [row['ESG_2021_rel_date']],
                'ESG_2022_rel_date': [row['ESG_2022_rel_date']],
                'ESG_2023_rel_date': [row['ESG_2023_rel_date']]
            }
            df_each_row = pd.DataFrame(each_row)
            df_result = pd.concat([df_result, df_each_row], ignore_index=True)
        else:
            for col in columns[1:]:
                if pd.isna(row[col]) is False:
                    if pd.isna(df_result[col].loc[df_result['stock_code'] == row['stock_code']].values[0]) is True:
                        df_result.loc[df_result['stock_code'] == row['stock_code'], col] = row[col]
                    else:
                        new_date = datetime.strptime(row[col], '%d/%m/%Y').date()
                        old_date = datetime.strptime(df_result[col].loc[df_result['stock_code'] == row['stock_code']].values[0], '%d/%m/%Y').date()
                        if new_date > old_date:
                            df_result.loc[df_result['stock_code'] == row['stock_code'], col] = row[col]
                else:
                    continue
    return df_result
```

```python
columns = ['stock_code', 'ESG_2017_rel_date', 'ESG_2018_rel_date', 'ESG_2019_rel_date', 'ESG_2020_rel_date', 'ESG_2021_rel_date', 'ESG_2022_rel_date', 'ESG_2023_rel_date']
df_result = pd.DataFrame(columns=columns)

df_result = concatenateData(df_result, df_2017, columns)
df_result = concatenateData(df_result, df_2018, columns)
df_result = concatenateData(df_result, df_2019, columns)
df_result = concatenateData(df_result, df_2020, columns)
df_result = concatenateData(df_result, df_2021, columns)
df_result = concatenateData(df_result, df_2022, columns)
df_result = concatenateData(df_result, df_2023, columns)

# df_result = df_result.sort_values(by='release_date')
df_result
```
